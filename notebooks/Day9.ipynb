### Part 1

```python
l = [int(i) for i in user_input]
```

```python
disk = []
i = 0
for i, n in enumerate(l):
    if i%2 == 0:
        disk += n*[i//2]
    else:
        disk += n*['.']
```

```python
space_idx = [i for i, x in enumerate(disk) if x == '.']
space_idx_rev = space_idx[::-1]
```

```python
while len(disk) > space_idx_rev[-1]:
    this = disk.pop()
    if this != '.':
        idx = space_idx_rev.pop()
        disk[idx] = this
```

```python
sum([i*d for i, d in enumerate(disk)])
```

```python
def blocks_to_string(blocks):
    end = ''
    for block in blocks:
        end += block['file_length']*str(block['idx'])
        end += block['spaces']*'.'
    return end
```

```python
def blocks_to_list(blocks):
    end = []
    for block in blocks:
        end += block['file_length']*[block['idx']]
        end += block['spaces']*['.']
    return end
```

```python
blocks = []
for i, (block, spaces) in enumerate(np.reshape(l, (len(l)//2, 2))):
    blocks.append({'file_length': int(block), 
                   'spaces': int(spaces), 
                   'idx': i})

moved_idx = []
fixed = []

while len(blocks) > 0:
    end_block = blocks.pop()

    if len(blocks) == 0:
        fixed.append(end_block)
        continue

    if end_block['idx'] in moved_idx:
        fixed.append(end_block)
        continue

    for i, b in enumerate(blocks):
        if end_block['file_length'] <= b['spaces']:
            blocks[-1]['spaces'] += end_block['file_length'] + end_block['spaces']
            end_block['spaces'] = b['spaces'] - end_block['file_length']
            b['spaces'] = 0
            blocks.insert(i+1, end_block)
            moved_idx.append(end_block['idx'])
            break
    else:
        fixed.append(end_block)

fixed = fixed[::-1]
```

```python
sum([i*b for i, b in enumerate(blocks_to_list(fixed)) if b != '.'])
```
